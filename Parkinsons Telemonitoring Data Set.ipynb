```python
# Modeling
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.NHR, train['Jitter(Abs)'], color='blue')
plt.xlabel("NHR")
plt.ylabel("Jitter(Abs)")
plt.show()
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['NHR']])
train_y = np.asanyarray(train[['Jitter(%)']])

regr.fit (train_x, train_y)


# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
# Plot outputs
```

```python
plt.scatter(train['NHR'], train['Jitter(%)'], color='red')

plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')

plt.xlabel('NHR')
plt.ylabel('Jitter(%)')

plt.show()
```

```python
# Evaluation of the linear model
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['NHR']])
test_y = np.asanyarray(test[['Jitter(%)']])
test_y_ =regr.predict(test_x)

print("MAE: %2f"% np.mean(np.absolute(test_y_-test_y)))
print("MSE: %2f"% np.mean((test_y_-test_y)**2))
print("R2-score: %2f"% r2_score(test_y, test_y_))
```
